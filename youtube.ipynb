```python
import subprocess

# ================================
# URL КАНАЛА — ПО УМОЛЧАНИЮ GenesisRidvan
# ================================
# Если нужно снова запрашивать канал вручную — раскомментируйте 2 строчки ниже:
# channel_url = input("Введите URL YouTube канала: ").strip()
channel_url = "https://www.youtube.com/@GenesisRidvan"

print("Используется канал:", channel_url)
print("Получаю список всех видео и стримов...")

# ================================
# ПОЛУЧЕНИЕ ВСЕХ ВИДЕО И СТРИМОВ
# ================================
result = subprocess.run(
    ["yt-dlp", "--flat-playlist", "--print", "%(id)s", channel_url],
    stdout=subprocess.PIPE,
    text=True
)
video_ids = [v for v in result.stdout.strip().split("\n") if v]

print(f"Найдено видео: {len(video_ids)}")

# ================================
# ФОРМИРОВАНИЕ СТРОКИ ДЛЯ run.ts
# ================================
joined_ids = "|".join(video_ids)
print("\nСтрока передаётся в run.ts:")
print(joined_ids[:500] + "...")  # выводим начало

# ================================
# ЗАПУСК СКРИПТА
# ================================
print("\nЗапуск run.ts...\n")
#!xvfb-run -s "-screen 0 1920x1080x24" npx ts-node src/run.ts "$joined_ids"
!npx ts-node src/run.ts "$joined_ids"
```
